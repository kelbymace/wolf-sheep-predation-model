```python
from wolf_sheep_rl import WolfSheepModel, PolicyNetwork, animate_model
import torch
import matplotlib.pyplot as plt
from IPython.display import HTML
import os
```

```python
policy_path = "../policies"
rl_policy = "best_sheep_policy_1.7.pt"

policy_net = PolicyNetwork(input_dim=51, hidden_dim=32, num_actions=8)
policy_net.load_state_dict(torch.load(os.path.join(policy_path, rl_policy)))
policy_net.eval()
```

```python
model = WolfSheepModel(
    width=30,
    height=30,
    initial_number_sheep=1,
    initial_number_wolves=5,
    model_version="sheep-wolves-grass",
    sheep_strategy="rl",
    wolf_strategy="seek_sheep",
    enable_grass=True,
    grass_regrowth_time=30,
    sheep_sight_radius=2,
    wolf_sight_radius=3,
    policy_net=policy_net,
)

model.collect_log_probs = False
model.setup()

anim = animate_model(model, steps=50, interval=120)
plt.close(anim._fig)
HTML(anim.to_jshtml())
```
